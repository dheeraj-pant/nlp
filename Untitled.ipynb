```python
import pandas as pd
dataset = pd.read_csv('train data.csv')
dataset.head()
```

```python
test_data= pd.read_csv('test data.csv')
test_data.head()
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(26,8.27)})
sns.countplot(x='class',data=dataset)
```

```python
import spacy
import re
nlp = spacy.load('en_core_web_sm')
  
# Create a Doc object
stopword_set=spacy.lang.en.stop_words
corpus = []
for i in range(0,len(dataset['text'])):
    doc = re.sub('[^a-zA-Z]', ' ', dataset['text'][i])
    doc = doc.lower()
    doc = nlp(doc)
    doc = " ".join([token.lemma_ for token in doc if not (nlp.vocab[token.text].is_stop  or token.text.startswith(' '))])
    corpus.append(doc)
```

```python
nlp = spacy.load('en_core_web_sm')
  
# Create a Doc object
stopword_set=spacy.lang.en.stop_words
corpus_test = []
for i in range(0,len(test_data['text'])):
    doc_test = re.sub('[^a-zA-Z]', ' ', test_data['text'][i])
    doc_test = doc_test.lower()
    doc_test = nlp(doc_test)
    doc_test = " ".join([token.lemma_ for token in doc_test if not (nlp.vocab[token.text].is_stop  or token.text.startswith(' '))])
    corpus_test.append(doc_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
# Creating the Bag of Words model

X = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, 1].values

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0,stratify=y)

from sklearn.svm import SVC
classifier= SVC(kernel= 'linear')
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
